## Detecting Outliers

First we have to see the distribution of the feature Station...

We can see obviously that station 29 has a peak value of 2000...


```python
sns.distplot(df.STATION_29)
```

```python
sns.boxplot(df.STATION_29)
```

We can see that there are certainly outliers in this station.

We are going to see how many outlier there are.

```python
upper_lim = df['STATION_29'].quantile(.95)
lower_lim = df['STATION_29'].quantile(.05)

temp = df[(df['STATION_29'] > upper_lim) | (df['STATION_29'] < lower_lim)]
temp.describe()
```

We can see that less than 10 percent of the data is the outlier.

Therefore we decided to drop it at.

```python
#Capping the outlier rows with Percentiles
upper_lim = df['STATION_29'].quantile(.95)
lower_lim = df['STATION_29'].quantile(.05)

df_2 = df[(df['STATION_29'] < upper_lim) & (df['STATION_29'] > lower_lim)]
```

```python
df_2.describe()
```

```python
df_2.Fail.value_counts().plot(kind="bar", color=["salmon", "lightblue"])
```

```python
df.Fail.value_counts()
```

## Explore the Correlation

We are going to see which station has the most influence with the failure rate

```python
df_2.drop('Fail', axis=1).corrwith(df_2.Fail).plot(kind='bar', grid=True, figsize=(16, 12), title="Correlation with Fail")
```

We can clearly see that Station 5 ,21 25, 28 is positively correlated the Failure.

And, Station 7, 13 ,15 is negatively correlated with Failure

We will scatter plot data from station 7, 21 

```python
# Create another figure
plt.figure(figsize=(10, 8))

plt.scatter(df_2.STATION_21[df_2.Fail==1],
            df_2.STATION_7[df_2.Fail==1],
            c="salmon")

plt.scatter(df_2.STATION_21[df_2.Fail==0],
            df_2.STATION_7[df_2.Fail==0],
            c="lightblue")

# Add some helpful info
plt.title("Station 21 vs Station 7")
plt.xlabel("Station 21")
plt.ylabel("Station 7")
plt.grid(b=True)
plt.legend(["Fail", "Not Fail"]);
```

As you can see, since Station 7 is negatively correlated with Failure, the instance with high Station 7 value tend to fail less

## See the data in the form of time series

We want to see the number of product failure in respect to date, month, year.

```python
# Convert Column 1 to date time index
df_3 = df_2.copy()
df_3 = df_3.set_index(pd.DatetimeIndex(df_3['MEASURE_DATETIME'])).groupby(pd.Grouper(freq='d')).mean().dropna(how='all')
```

```python
df_3.head(10)
```

```python
import calmap

fig, ax = calmap.calendarplot(df_3['Fail'], fillcolor='grey', cmap='YlOrRd', fig_kws=dict(figsize=(100,50)))
fig.colorbar(ax[0].get_children()[1], ax=ax.ravel().tolist())
```

### Conclusion

We can see that the failure mostly occur during the
* January and December of 2021
* November of 2030
* August and November of 2031
